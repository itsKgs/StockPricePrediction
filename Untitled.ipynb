### Outlier Detection using Z-Score Method

```python
# Define the threshold for identifying an outlier
threshold = 3
# Calculate the Z-scores for the log returns
df['Z_Score'] = np.abs((df['LogReturn'] - df['LogReturn'].mean()) / df['LogReturn'].std())
# Identify the outliers
outliers = df[df['Z_Score'] > threshold]
```

```python
print(f"\nFound {len(outliers)} days with a log return Z-Score greater than {threshold}.")
```

```python
if not outliers.empty:
    print("These are the dates with the most significant price swings (potential outliers):")
    # Display the outliers, sorted by the Z-score
    print(outliers.sort_values(by='Z_Score', ascending=False)[['Close', 'LogReturn', 'Z_Score']])
else:
    print("No significant outliers were detected based on the Z-score method.")
```

### Outlier Detection using a Rolling Window

```python
# Calculate rolling mean and std
window = 20  # 20-day rolling window
df['rolling_mean'] = df['Close'].rolling(window=window).mean()
df['rolling_std'] = df['Close'].rolling(window=window).std()

# Identify outliers (price deviates > 3*rolling_std from rolling_mean)
df['outlier'] = (abs(df['Close'] - df['rolling_mean']) > 3 * df['rolling_std'])

# Plot
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 8))
plt.plot(df['Close'], label='Close Price', alpha=0.8)
plt.plot(df['rolling_mean'], color='orange', linestyle='--', label='20-Day Rolling Mean')
plt.scatter(df.index[df['outlier']], df['Close'][df['outlier']], color='red', s=100, label='Outliers')
plt.legend()
plt.title('Apple Stock Price Outlier Detection using a Rolling Window')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Display the dates identified as outliers
outlier_dates = df[df['outlier']]
if not outlier_dates.empty:
    print("\nDates identified as outliers:")
    print(outlier_dates[['Close', 'rolling_mean', 'rolling_std']])
else:
    print("\nNo outliers were detected with this method.")
```

# Stationarity Test

```python
#Perform The Augemented Dickey-Fuller Test
def adf_Test(data):
    adf_test = adfuller(data)

    print("ADF Statistic: {:.4f}".format(adf_test[0]))
    print("p-value: {:.4f}".format(adf_test[1]))

    print("Critical Values:")
    for key, value in adf_test[4].items():
        print(f"   {key}: {value:.4f}")

    # Interpretation
    if adf_test[1] < 0.05:
        print("\nThe time series is likely stationary.")
    else:
        print("\nThe time series is likely non-stationary.")
```

```python
# Performs and interprets the Kwiatkowski-Phillips-Schmidt-Shin Test
def kp_ss_Test(data):
    kpss_stat, p_value, lags, critical_values = kpss(data, regression='c', nlags="auto")
    
    print("KPSS Statistic: {:.4f}".format(kpss_stat))
    print("p-value: {:.4f}".format(p_value))

    for key, value in critical_values.items():
        print(f"   {key}: {value:.4f}")

    # Interpretation
    if p_value < 0.05:
        print("\nThe time series is likely non-stationary.")
    else:
        print("\nThe time series is likely stationary.")
```

```python
adf_Test(df["Close"])
```

```python
kp_ss_Test(df["Close"])
```

### Differencing

```python
# Apply First-Order Differencing
df['log_diff'] = df['log_Close'].diff()
```

```python
adf_Test(df['log_Close'].diff().dropna())
```

```python
kp_ss_Test(df['log_Close'].diff().dropna())
```

```python
plt.figure(figsize=(12,6))
plt.title('Stationary Series (Log-Differenced Closing Price)')
plt.plot(df['log_diff'])
plt.axhline(y=0, linestyle='--', color='red') # Add a zero line
plt.grid(True)
plt.show()
```
